```python
import matplotlib.pyplot as plt
import numpy as np

# Example rainfall data for 5 years (60 months)
months = np.arange(1, 61)
rainfall = np.random.normal(loc=100, scale=50, size=60)  # Random rainfall data

plt.figure(figsize=(10, 5))
plt.plot(months, rainfall, label='Rainfall (in mm)', color='b')
plt.title("Monthly Rainfall Over 5 Years")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.grid(True)
plt.show()
```
