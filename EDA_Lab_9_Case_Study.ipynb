### **Module 1: Introduction to EDA**

**1. Measurement Scales Identification**

* **`movie_id`**: **Nominal** (Identifier only; math like averaging ID numbers is meaningless).
* **`genre`**: **Nominal** (Categories with no inherent mathematical order).
* **`rating`**: **Ordinal** (Ordered categories where 5 > 4, but the "distance" between them is subjective) or **Interval** if treated as a score.
* **`age`**: **Ratio** (Has a true zero point and mathematical operations are valid).


```
import pandas as pd
df = pd.read_csv('movie_ratings_dirty.csv')

print(df.head(10))

# Count missing values
null_counts = df.isnull().sum()
# Calculate percentage for rating
rating_null_pct = (df['rating'].isnull().sum() / len(df)) * 100

print(null_counts)
print(f"Missing Rating Percentage: {rating_null_pct:.2f}%")
```

This code snippet loads the `movie_ratings_dirty.csv` file into a pandas DataFrame named `df`. It then prints the first 10 rows of the DataFrame to give an initial view of the data structure and content. Following this, it calculates and displays the number of missing values (`NaN`) in each column using `df.isnull().sum()`. Finally, it specifically calculates the percentage of missing values in the `rating` column, which is crucial for understanding data completeness for a key analytical feature.

**3. Detecting Anomalies via Descriptive Statistics**

```
# Identify outliers like age=150 or rating=10
print(df[['age', 'rating']].describe())
```

This code applies the `describe()` method to the `age` and `rating` columns. The output provides key descriptive statistics such as count, mean, standard deviation, minimum, maximum, and quartile values. This helps in identifying potential data anomalies, such as an `age` of 150 (which is unrealistic) or a `rating` of 10 (which might be outside an assumed scale of 1-5 or 1-7 depending on the dataset's context), indicating the need for data cleaning.

**4. Data Refactoring Significance**
Refactoring ensures that identifiers (`user_id`) are treated as objects/strings rather than integers. This prevents the model or statistical summary from calculating meaningless metrics like the "average user ID."



```
df['movie_id'] = df['movie_id'].astype(str)
df['user_id'] = df['user_id'].astype(str)
print(df[['movie_id', 'user_id']].dtypes)
```

This code performs **data refactoring** by converting the `movie_id` and `user_id` columns to `object` (string) data types. This is important because these columns are unique identifiers and should not be treated as numerical data where statistical operations like calculating a mean would be nonsensical. By converting them to strings, we ensure they are handled appropriately for identification and categorical analysis, aligning with the principles of data refactoring for proper statistical analysis.

**5. Dropping Columns**

```
# Drop director for demographic study
df_reduced = df.drop(columns=['director'])
```

This code demonstrates how to drop a specific column from the DataFrame using the `drop()` method. In this example, the `director` column is dropped, creating a new DataFrame `df_reduced`. This action is typically performed when a particular column is not relevant for a specific analysis (e.g., a user-centric demographic study where the director's information is not needed), thereby simplifying the dataset and reducing computational overhead.

### **Module 2: Data Transformation**

**6. Data Deduplication**

```
print(f"Before: {len(df)}")
df = df.drop_duplicates()
print(f"After: {len(df)}")
```

This cell performs data deduplication. It first prints the number of rows in the DataFrame before deduplication using `len(df)`. Then, `df.drop_duplicates()` is applied to remove any rows that are exact duplicates, updating the DataFrame `df`. Finally, it prints the number of rows after deduplication to show how many duplicates were removed. Deduplication is an essential step in data cleaning to ensure data integrity and prevent biased analysis due to redundant entries.

**7. Merging Dataframes**

To demonstrate how a combined DataFrame like our `df` is constructed from separate entities, let's simulate having distinct `df_users`, `df_movies`, and `df_reviews` DataFrames. We'll extract the relevant columns from our existing `df` to ensure consistency in data and column names.

```
# 1. Create a hypothetical df_users DataFrame
# Contains unique user information (user_id, age, gender)
df_users_reconstructed = df[['user_id', 'age', 'gender']].drop_duplicates().reset_index(drop=True)

# 2. Create a hypothetical df_movies DataFrame
# Contains unique movie information (movie_id, title, genre, director, release_year)
df_movies_reconstructed = df[['movie_id', 'title', 'genre', 'director', 'release_year']].drop_duplicates().reset_index(drop=True)

# 3. Create a hypothetical df_reviews DataFrame (the interaction table)
# Contains user-movie interactions and their specific rating/review (user_id, movie_id, rating, review_text)
df_reviews_reconstructed = df[['user_id', 'movie_id', 'rating', 'review_text']].copy()

print("Reconstructed df_users (sample):")
display(df_users_reconstructed.head())
print("\nReconstructed df_movies (sample):")
display(df_movies_reconstructed.head())
print("\nReconstructed df_reviews (sample):")
display(df_reviews_reconstructed.head())
```

Now that we have our separate DataFrames, we can demonstrate how they would be merged to form a structure similar to our original `df`. We'll typically start by merging the user information with the reviews, and then add the movie details.

```
# Step A: Merge df_users_reconstructed with df_reviews_reconstructed on 'user_id'
# This combines user demographic data with their reviews and ratings.
users_reviews_merged = pd.merge(
    df_users_reconstructed,
    df_reviews_reconstructed,
    on='user_id',
    how='inner' # Use inner join to keep only users who have left reviews
)

print("Users and Reviews Merged (sample):")
display(users_reviews_merged.head())
```

The previous step combined user demographics with their specific reviews and ratings. Now, we'll merge this result with the movie details to get a comprehensive view, similar to our original `df`.

```
# Step B: Merge the combined users_reviews_merged with df_movies_reconstructed on 'movie_id'
# This adds the movie's title, genre, director, and release year to each review entry.
final_reconstructed_df = pd.merge(
    users_reviews_merged,
    df_movies_reconstructed,
    on='movie_id',
    how='inner' # Use inner join to keep only reviews for existing movies
)

# Reorder columns to closely match the original df's structure for comparison
final_reconstructed_df = final_reconstructed_df[[
    'user_id', 'age', 'gender', 'movie_id', 'title', 'genre', 'director',
    'release_year', 'rating', 'review_text'
]]

print("Final Reconstructed DataFrame (sample, matching original df structure):")
display(final_reconstructed_df.head())
```

This `final_reconstructed_df` now closely mirrors the structure and content of our initial `df`, demonstrating how a single comprehensive DataFrame can be built by merging information from various logically separate tables representing users, movies, and their interactions.

```
# Assuming you have separate df_users and df_movies
# df_merged = pd.merge(df_users, df_ratings, on='user_id', how='inner')
```

```
# Create sample df_users and df_movies DataFrames for demonstration with more attributes
df_users = pd.DataFrame({
    'user_id': ['U27', 'U21', 'U20', 'U29', 'U100'],
    'age': [63, 32, 31, 62, 25],
    'gender': ['Male', 'Non-binary', 'Female', 'Male', 'Female'],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami'],
    'occupation': ['Engineer', 'Artist', 'Teacher', 'Retired', 'Student'],
    'membership_level': ['Gold', 'Silver', 'Gold', 'Bronze', 'Silver']
})

df_movies = pd.DataFrame({
    'movie_id': ['101', '104', '112', '103', '200'],
    'title': ['Inception', 'Mad Max: Fury Road', 'Everything Everywhere All At Once', 'Parasite', 'New Movie'],
    'director': ['Christopher Nolan', 'George Miller', 'Daniels', 'Bong Joon-ho', 'New Director'],
    'production_company': ['Warner Bros', 'Warner Bros', 'A24', 'CJ Entertainment', 'Indie Prod'],
    'critic_score': [8.8, 8.1, 8.5, 9.0, 7.5]
})

# Create a simplified df_ratings for demonstration, linking users and movies
df_ratings = df[['user_id', 'movie_id', 'rating']].copy()

print("Sample df_users (with more attributes):")
display(df_users)
print("\nSample df_movies (with more attributes):")
display(df_movies)
print("\nSample df_ratings (from original df):")
display(df_ratings.head())
```

First, let's demonstrate an **Inner Join**. An inner join returns only the rows that have matching values in both DataFrames being joined. In this case, it will combine user demographics with their ratings only for users and movies that exist in *both* `df_users` and `df_ratings`.

```
# Inner Join: Combine df_users with df_ratings based on user_id
inner_joined_df = pd.merge(df_users, df_ratings, on='user_id', how='inner')

print("Inner Joined DataFrame (df_users + df_ratings):")
display(inner_joined_df.head())
```

Next, let's demonstrate a **Left Join**. A left join returns all rows from the left DataFrame (`df_users` in this case) and the matching rows from the right DataFrame (`df_ratings`). If there's no match in the right DataFrame, it fills with `NaN` values for the columns from the right DataFrame. This is useful when you want to keep all user information, even if they haven't provided any ratings in `df_ratings`.

```
# Left Join: Combine df_users with df_ratings based on user_id
left_joined_df = pd.merge(df_users, df_ratings, on='user_id', how='left')

print("Left Joined DataFrame (df_users + df_ratings):")
display(left_joined_df.head())
```

You can also join the `df_movies` with the `inner_joined_df` or `left_joined_df` to bring in movie details based on `movie_id`. Here's an example using the `left_joined_df` to include movie titles and directors for all rated movies, keeping all user and rating information:


```
# Further join with df_movies to add movie details
final_merged_df = pd.merge(left_joined_df, df_movies, on='movie_id', how='left')

print("Final Merged DataFrame (Users + Ratings + Movies using Left Join):")
display(final_merged_df.head())
```

This commented-out code snippet illustrates how one would merge two separate DataFrames (`df_users` and `df_ratings`) using pandas' `merge()` function. It shows an example of an `inner join` based on the common `user_id` column. Merging is crucial when data is distributed across multiple tables and needs to be combined for comprehensive analysis, such as linking user demographics with their movie ratings.

**8. Handling Missing Ratings**

```
# (a) Drop
df_dropped = df.dropna(subset=['rating'])
# (b) Mean Imputation
df_mean = df.copy()
df_mean['rating'] = df_mean['rating'].fillna(df['rating'].mean())
# (c) Forward Fill
df_ffill = df['rating'].fillna(method='ffill')
```

This code demonstrates three different techniques for handling missing values in the `rating` column:  
1.  **Dropping rows**: `df_dropped = df.dropna(subset=['rating'])` creates a new DataFrame where all rows with missing `rating` values are removed.  
2.  **Mean Imputation**: `df_mean['rating'] = df_mean['rating'].fillna(df['rating'].mean())` replaces missing `rating` values with the mean of the existing `rating` values.  
3.  **Forward Fill**: `df_ffill = df['rating'].fillna(method='ffill')` fills missing `rating` values with the last valid observation in the column. Each method has different implications for the dataset and its subsequent analysis.

**9. Categorical Standardization**

```
df['genre'] = df['genre'].replace({'sci-fi': 'Sci-Fi'})
df['gender'] = df['gender'].replace({'M': 'Male'})
```

This code performs standardization on categorical columns to ensure consistency in data entry. It uses the `replace()` method to:  
-   Convert variations of 'sci-fi' to 'Sci-Fi' in the `genre` column.  
-   Convert 'M' to 'Male' in the `gender` column.  
This process is vital for accurate categorical analysis and visualization, as inconsistent spellings or abbreviations can lead to categories being treated as distinct when they should be unified.

**10. Discretization and Binning**

```
bins = [0, 19, 35, 60, 100]
labels = ['Teen', 'Young Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

This code demonstrates **discretization** by creating a new categorical column `age_group` from the numerical `age` column. It defines `bins` (age ranges) and corresponding `labels` ('Teen', 'Young Adult', 'Middle Aged', 'Senior'). The `pd.cut()` function then assigns each `age` value to its appropriate age group. This technique transforms continuous numerical data into discrete categories, which can be useful for analyses requiring grouped age data or for simplifying complex numerical distributions.

**11. Outlier Filtering (IQR Method)**

```
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
df_clean = df[~((df['age'] < (Q1 - 1.5 * IQR)) | (df['age'] > (Q3 + 1.5 * IQR)))]
```

**Addressing Inconsistencies for User Profiles**

Based on your observations, we need to correct specific inconsistencies in the `df` before creating user-centric DataFrames like `df_users_reconstructed`. This involves:
1.  **Correcting Anomalous Age:** `user_id` U27 has an age of 150, which is an outlier. We will standardize this to 63, which is also present for this user and more realistic.
2.  **Standardizing Gender:** `user_id` U21 appears with both 'Non-binary' and 'Male' genders. For consistency in a user profile, we will standardize all entries for U21 to 'Non-binary'.

```
# 1. Correct Age for user_id 'U27': Replace 150 with 63
df.loc[(df['user_id'] == 'U27') & (df['age'] == 150), 'age'] = 63

# 2. Standardize Gender for user_id 'U21': Set to 'Non-binary'
df.loc[df['user_id'] == 'U21', 'gender'] = 'Non-binary'

print("DataFrame 'df' updated for U27 age and U21 gender inconsistencies.")
```

Now that the `df` has been corrected for these specific user profile inconsistencies, we will re-generate `df_users_reconstructed` to reflect these changes. The `df_movies_reconstructed` and `df_reviews_reconstructed` do not need to be re-generated as the corrections were user-specific.

```
# Re-create df_users_reconstructed with corrected df
df_users_reconstructed = df[['user_id', 'age', 'gender']].drop_duplicates().reset_index(drop=True)

print("Reconstructed df_users (sample) after corrections:")
display(df_users_reconstructed.head())
```

This code demonstrates **outlier detection and filtering** using the Interquartile Range (IQR) method for the `age` column.  
-   It calculates the first quartile (Q1) and third quartile (Q3) of the `age` data.  
-   The IQR is then computed as the difference between Q3 and Q1.  
-   Outliers are identified as values falling below `Q1 - 1.5 * IQR` or above `Q3 + 1.5 * IQR`.  
-   Finally, `df_clean` is created, containing only the rows where `age` is within the calculated acceptable range, effectively removing outliers like the '150-year-old user'.

### **Module 3: Grouping and Correlation**

**12. Groupby Aggregation**

```
director_ranking = df.groupby('director')['rating'].mean().rename("Average_Score").sort_values(ascending=False)
print(director_ranking.head(3))
```

This code uses the `groupby()` method to calculate the average `rating` for each `director`.  
-   `df.groupby('director')['rating'].mean()` groups the DataFrame by director and then calculates the mean rating for each group.  
-   `.rename("Average_Score")` renames the resulting series to `Average_Score`.  
-   `.sort_values(ascending=False)` sorts the directors by their average score in descending order.  
-   `print(director_ranking.head(3))` displays the top 3 directors based on this average rating. This helps in identifying the most critically acclaimed directors according to user ratings.

**13. Pivot Table Analysis**

```
pivot = df.pivot_table(index='genre', columns='gender', values='rating', aggfunc='mean')
print(pivot)
```

This code creates a **pivot table** to analyze the average `rating` across different `genre` and `gender` combinations.  
-   `df.pivot_table()` is used with `index='genre'` (rows), `columns='gender'` (columns), `values='rating'` (data to aggregate), and `aggfunc='mean'` (aggregation function).  
-   The resulting pivot table `pivot` shows the average rating for each genre, broken down by gender, providing insights into genre-specific audience preferences across different gender groups.

**14. Types of Analysis Examples**

To better understand the differences, let's look at examples using our movie dataset:

*   **Univariate Analysis:** This involves analyzing a single variable. For instance, creating a histogram to visualize the distribution of `ratings` to understand how frequently each rating appears. This helps in grasping the central tendency and spread of individual features.
    *   _Example:_ Histogram of `ratings`.

*   **Bivariate Analysis:** This explores the relationship between two variables. A scatter plot of `age` vs. `rating` can show if there's any pattern or correlation between a user's age and the ratings they give. This helps in understanding how two variables move together.
    *   _Example:_ Scatter plot of `age` vs `rating`.

*   **Multivariate Analysis:** This examines the relationships among three or more variables simultaneously. A heatmap showing the correlation between `age`, `release_year`, and `rating` could reveal complex interactions, such as how both age and release year might influence ratings. This is useful for uncovering more intricate patterns within the data.
    *   _Example:_ Heatmap showing correlation between `age`, `release_year`, and `rating`.

```
# Feature engineering
df['review_length'] = df['review_text'].str.split().str.len()

# Correlation (ignoring NaNs)
correlation = df['review_length'].corr(df['rating'])
print(f"Correlation: {correlation}")
```

This code performs a **bivariate analysis** to understand the relationship between review length and movie rating.  
-   **Feature Engineering**: A new column `review_length` is created by calculating the word count of each `review_text` using `df['review_text'].str.split().str.len()`.  
-   **Correlation Calculation**: `df['review_length'].corr(df['rating'])` calculates the Pearson correlation coefficient between `review_length` and `rating`, ignoring any NaN values.  
-   The output `Correlation: 0.2678157688184501` suggests a weak positive correlation, indicating that slightly longer reviews tend to be associated with slightly higher ratings, although the relationship is not strong.
