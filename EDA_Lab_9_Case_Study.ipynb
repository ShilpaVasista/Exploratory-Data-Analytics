## Solutions for EDA & Data Transformation Case Study

### **Module 1: Introduction to EDA**

**1. Measurement Scales Identification**

* **`movie_id`**: **Nominal** (Identifier only; math like averaging ID numbers is meaningless).
* **`genre`**: **Nominal** (Categories with no inherent mathematical order).
* **`rating`**: **Ordinal** (Ordered categories where 5 > 4, but the "distance" between them is subjective) or **Interval** if treated as a score.
* **`age`**: **Ratio** (Has a true zero point and mathematical operations are valid).


```
import pandas as pd
df = pd.read_csv('movie_ratings_dirty.csv')

# Count missing values
null_counts = df.isnull().sum()
# Calculate percentage for rating
rating_null_pct = (df['rating'].isnull().sum() / len(df)) * 100

print(null_counts)
print(f"Missing Rating Percentage: {rating_null_pct:.2f}%")
```

**3. Detecting Anomalies via Descriptive Statistics**

```
# Identify outliers like age=150 or rating=10
print(df[['age', 'rating']].describe())
```

**4. Data Refactoring Significance**
Refactoring ensures that identifiers (`user_id`) are treated as objects/strings rather than integers. This prevents the model or statistical summary from calculating meaningless metrics like the "average user ID."

**5. Dropping Columns**

```
# Drop director for demographic study
df_reduced = df.drop(columns=['director'])
```

### **Module 2: Data Transformation**

**6. Data Deduplication**

```
print(f"Before: {len(df)}")
df = df.drop_duplicates()
print(f"After: {len(df)}")
```

**7. Merging Dataframes**

```
# Assuming you have separate df_users and df_movies
# df_merged = pd.merge(df_users, df_ratings, on='user_id', how='inner')
```

**8. Handling Missing Ratings**

```
# (a) Drop
df_dropped = df.dropna(subset=['rating'])
# (b) Mean Imputation
df_mean = df.copy()
df_mean['rating'] = df_mean['rating'].fillna(df['rating'].mean())
# (c) Forward Fill
df_ffill = df['rating'].fillna(method='ffill')
```

**9. Categorical Standardization**

```
df['genre'] = df['genre'].replace({'sci-fi': 'Sci-Fi'})
df['gender'] = df['gender'].replace({'M': 'Male'})
```

**10. Discretization and Binning**

```
bins = [0, 19, 35, 60, 100]
labels = ['Teen', 'Young Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

**11. Outlier Filtering (IQR Method)**

```
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
df_clean = df[~((df['age'] < (Q1 - 1.5 * IQR)) | (df['age'] > (Q3 + 1.5 * IQR)))]
```

### **Module 3: Grouping and Correlation**

**12. Groupby Aggregation**

```
director_ranking = df.groupby('director')['rating'].mean().rename("Average_Score").sort_values(ascending=False)
print(director_ranking.head(3))
```

**13. Pivot Table Analysis**

```
pivot = df.pivot_table(index='genre', columns='gender', values='rating', aggfunc='mean')
print(pivot)
```
