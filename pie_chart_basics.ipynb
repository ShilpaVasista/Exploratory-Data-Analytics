Let's cover each `matplotlib.pyplot.pie()` parameter with a one-line explanation and a simple code modification.

**1. Labels:**

* **Explanation:** Assigns text labels to each wedge of the pie chart.
* **Code Modification:**

```
import matplotlib.pyplot as plt

data = [30, 20, 50]
labels = ['Category X', 'Category Y', 'Category Z']
plt.pie(data, labels=labels)
plt.show()
```

**2. Colors:**

* **Explanation:** Sets the color of each wedge in the pie chart.
* **Code Modification:**

```
import matplotlib.pyplot as plt

data = [30, 20, 50]
colors = ['yellow', 'purple', 'orange']
plt.pie(data, colors=colors)
plt.show()
```

**3. Hatch Slices:**

* **Explanation:** Adds patterned fills (hatches) to the wedges for visual distinction.
* **Code Modification:**

```
import matplotlib.pyplot as plt
#hatch must consist of a string of *+-./OX\ox|
data = [30, 20, 50]
hatches = ['/', '*', 'o'] #original hatches]
plt.pie(data, hatch=hatches)
plt.show()
hatches = ['x', '.', '-'] #modified hatches.
plt.pie(data, hatch=hatches)
plt.show()
```

**4. Autopct:**

* **Explanation:** Displays the percentage of each wedge on the pie chart.
* **Code Modification:**

```
import matplotlib.pyplot as plt

data = [30, 20, 50]
plt.pie(data, autopct='%1.1f%%')
plt.show()
```

**5. Explode:**

* **Explanation:** Offsets a wedge from the center of the pie chart for emphasis.
* **Code Modification:**

```
import matplotlib.pyplot as plt

data = [30, 20, 50]
explode = [0.1, 0, 0] #modified explode. first slice offset
plt.pie(data, explode=explode)
plt.show()
```

**6. Startangle:**

* **Explanation:** Rotates the starting point of the pie chart.
* **Code Modification:**

```
import matplotlib.pyplot as plt

data = [30, 20, 50]
plt.pie(data, startangle=0) #original startangle
plt.show()
plt.pie(data, startangle=140) #modified startangle, rotated 140 degrees
plt.show()
```
