# Outlier Detection and Filtering

 Loading the Dataset

```python
# Import required libraries
import pandas as pd

# Load the dataset from GitHub
df = pd.read_csv('sales.csv')

# Display the first 10 rows of the dataset
df.head(10)
```

Adding a New Column: TotalPrice
Code Block: Calculate the TotalPrice for each transaction.

```python
# Add a new column 'TotalPrice'
df['TotalPrice'] = df['UnitPrice'] * df['Quantity']

# Display the dataset with the new column
df.head()
```

Detecting Outliers

Code Block: Identify transactions where TotalPrice > 3,000,000.

```python
# Detect outliers where TotalPrice > 3,000,000
outliers = df[df['TotalPrice'] > 3000000]

# Display the outliers
outliers
```

Displaying Rows with Extreme Outliers

Objective: Display rows where TotalPrice > 6,741,112.

```python
# Detect extreme outliers where TotalPrice > 6,741,112
extreme_outliers = df[df['TotalPrice'] > 6741112]

# Display extreme outliers
extreme_outliers
```
