Python Program with Heatmap using Iris Dataset:
- Import libraries and Load data “iris.csv”
- Calculate the correlation matrix using iris.corr()
- Create a figure and use plt.imshow to display the correlation as a heatmap with a
coolwarm colormap.

1. Data Loading and Exploration
- Import the necessary libraries: pandas, matplotlib.pyplot, and seaborn.
- Load the iris.csv dataset into a Pandas DataFrame and display the first few rows to understand its structure.
- Check for missing values and summarize the dataset's statistical properties.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Load the Iris dataset
df = pd.read_csv('Iris.csv')
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())
random_sample_pandas = df.sample(n=10)
print("\nRandom sample using pandas sample():")
print(random_sample_pandas)
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
# Display summary statistics
print("\nSummary statistics:")
print(df.describe())
```

2. Feature Correlation Analysis
- Compute the correlation matrix using df.corr() to analyze relationships between numerical features.
- Display the correlation matrix and identify highly correlated pairs.

```python
# Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Display the correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

3. Heatmap Visualization
- Create a heatmap using Seaborn’s heatmap function to visualize the correlation matrix with a color scale.
- Add annotations to show correlation coefficients on the heatmap and use the coolwarm colormap for better interpretation.

```python
# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Iris Dataset Correlation Matrix')
plt.show()
```

Python Program with Histogram using Iris Dataset:
- For the above dataset(Iris dataset), Create a figure and use plt.hist to visualize the
distribution of Sepal Length using 10 bins and black edge color.
- Add labels, title, and grid for better readability

4. Data Distribution Analysis
- Load the iris.csv dataset and inspect its structure.
- Create a histogram to analyze the distribution of sepal_length using 10 bins and a black edge color.
- Add appropriate labels, a title, and a grid to improve readability.

```python
df = pd.read_csv('Iris.csv')  # Replace with the correct file path
print(df.columns)

# a. Create a histogram for Sepal Length
plt.figure(figsize=(8, 6))
plt.hist(df['SepalLengthCm'], bins=10, edgecolor='black', color='skyblue')


# Add labels, title, and grid
plt.xlabel('Sepal Length (cm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Sepal Length in Iris Dataset', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
